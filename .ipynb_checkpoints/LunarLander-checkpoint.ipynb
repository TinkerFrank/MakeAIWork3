```python
import gymnasium as gym

from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_vec_env
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create environment that keeps track of state and reward
env = gym.make('LunarLander-v2', render_mode="human") #default is discrete 

## LUNAR LANDER INFO ##
## environment: landing pad (0,0), left/right/main engine on=full throttle or off 
## initial state:  top center of the viewport, a random initial force applied to its center of mass.
## state: 8-dim vector (x,y, vx, vy, ang_v, left_leg_touchdown = False, right_leg_touchdown = False)
## reward: +/- distance landing pad, +/- slower/faster moving, - degrees of tilt, ++ for each leg touchdown, - side engine fired, - main engine fired
## landing gives 100 or -100 for crashing, over 200 point = a solution
## Termination: crashes = body in contact with moon, outside viewport x>1, not awake = no movvement 


#reset the environment to its initial state 
observation, info = env.reset()

#steps: state -> action -> next state + reward
for _ in range(20): # _ underscore used to signify its not an important/dummy variable
  # Generate action
  action = env.action_space.sample()
  print("Action taken", action)

  # Execute action in environment and get: next state (observation) + reward
  # + terminated (end goal reached?), truncated (timelimit for out of bounds) and info (additional data)
  observation, reward, terminated, truncated, info = env.step(action)

  # Game is terminated if land or crashed or truncated
  if terminated or truncated:
    # Reset the environment
    print('Environment is reset')
    observation, info = env.reset()

env.close()


```

```python
# Initiate environment
env = gym.make('LunarLander-v2', render_mode="human")

# Instantiate the agent 
# PPO Policy architecture (with some extra parameters for faster training)
# PPO = Proximal Policy Optimization: combination of Value-based (most rewarding action) and Policy Based (prob. distr. over actions)# 
# we use MultilayerPerceptron (MlP) because the input is a vector
# if we had frames as input we would use CnnPolicy

model = PPO(  # from stable_baselines3 import PPO
    policy='MlpPolicy',
    env=env,
    n_steps=1024,
    batch_size=64,
    n_epochs=4,
    gamma=0.999,
    gae_lambda=0.98,
    ent_coef=0.01,
    verbose=1
)


# # Train the agent
# model.learn(total_timesteps=100)

# Train the agent
total_timesteps = 1000
update_interval = 100  # Update the plot every 10 time steps

ep_rewards = []
obs = env.reset()


from IPython import display


for t in range(total_timesteps):
    action, _states = model.predict(observation, deterministic=True)
    observation, reward, terminated, truncated, info = env.step(action)
    ep_rewards.append(reward)
    if terminated or truncated:
        observation, info = env.reset()

    # Update the reward graph every 'update_interval' time steps
#    if t % update_interval == 0:
#         reward_line.set_xdata(range(len(ep_rewards)))
#         reward_line.set_ydata(ep_rewards)
#         plt.plot()  # Redraw the plot
#         plt.pause(0.001)  # Pause to update the plot
#         #display.clear_output(wait=True)
#         display.display(plt.gcf())

plt.figure()
reward_line, = plt.plot(ep_rewards)
plt.xlabel('Time Steps')
plt.ylabel('Reward')
plt.title('Reward over Time')

plt.ioff()
plt.show()

# Save the model
model_name = "ppo-LunarLander-v2"
model.save(model_name)
```

```python
#evaluate performance
# from stable_baselines3.common.evaluation import evaluate_policy

# When you evaluate your agent, you should not use your training environment but create an evaluation environment.
eval_env = gym.make(env_id, render_mode="rgb_array")


mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)

print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```
