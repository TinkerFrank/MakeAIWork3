<h3>Exercise 2</h3>

<h4>Gegeven</h4>
<p>Dobbelstenen <i>d0</i> en <i>d1</i> en hun gecombineerde uitkomstruimte <i>dice</i></p>

<h3>Gevraagd</h3>
<p>
    Geef de <u>kansmassa-functie</u> van de som van het aantal ogen van <i>d0</i> en <i>d1</i><br>
    <strong>TIP</strong>: Maak eerst een dictionary met daarin per alle uitkomsten
</p>

<h4>Oplossing</h4>

```python

probabilities = {}

total = len(resultspace.flatten())

for number in resultspace.flatten():
    if number in probabilities:
        probabilities[number] += 1/total
    else:
        probabilities[number] = 1/total

print(probabilities)
# check if adds up to a total of 1
print(sum(probabilities.values()))
```

```python
import matplotlib.pyplot as plt
plt.bar(list(probabilities.keys()), list(probabilities.values()))
```

```python
# Simulating rolling two dice
dice_1 = np.arange(1, 7)
dice_2 = np.arange(1, 7)
rolls = np.add.outer(dice_1, dice_2)

# Calculating probabilities
total_outcomes = rolls.size
unique, counts = np.unique(rolls, return_counts=True)
probabilities = counts / total_outcomes

# Creating the histogram
plt.bar(unique, probabilities)
plt.xlabel('Sum of Two Dice')
plt.ylabel('Probability')
plt.title('Probability Histogram of Two Dice Rolls')
plt.show()
```
